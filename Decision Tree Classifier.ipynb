Decision Tree Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
# Independent features
X= pd.DataFrame(iris['data'],columns=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'])
```

```python
# dependent features
y = iris['target']
```

```python
## train test plot
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

In decision tree, there is no requirement of standardization or normalisation as we can directly apply the algorithm

```python
## apply Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth = 2)
model.fit(X_train,y_train)
```

```python
## Visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

Gini value ranges from 0 to 0.5 and Entropy value ranges from 0 to 1

To prevent overfitting there are two ways :
1. Post Prunning - After constructing the entire tree and then pruning it. It is always good for smaller datasets.
2. Pre Prunning - We use many parameters as we hyperparameter tune it.

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test,y_pred)
print(cm)
cr = classification_report(y_test,y_pred)
print(cr)
```

# Decision Tree Classifier Prepruning And Hyperparameter Tuning

```python
param = {
    'criterion' : ['gini','entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3,4,5],
    'max_features' : ['auto', 'sqrt', 'log2']


}
```

```python
param
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
treemodel = DecisionTreeClassifier()
grid = GridSearchCV(treemodel,param_grid=param,cv=5, scoring = 'accuracy')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
cr = classification_report(y_test,y_pred)
print(cr)
ac = accuracy_score(y_test,y_pred)
print(ac)
```
